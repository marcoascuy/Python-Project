```python
import pandas as pd
import numpy as np
import matplotlib as mlt
import plotly.express as px
import plotly.graph_objects as go
```

```python
resultados = pd.read_csv('results.csv')
```

```python
resultados
```

```python
resultados = resultados[resultados['home_score']>=0]
```

```python
chile = resultados[(resultados['home_team']=='Chile') | (resultados['away_team']=='Chile')]
```

```python
chile
```

```python
def marcador(df):
    if df.loc[:,'home_score'] > df.loc[:,'away_score']:
        df['r'] = 'Home'
    elif df.loc[:,'away_score'] > df.loc[:,'away_score']:
        df['r'] = 'Away'
    else:
        df['r'] = 'Tie'
    
    
```

```python
chile.groupby('tournament')['date'].count()
```

```python
chile['home_score'] = chile['home_score'].astype('int64')
chile['away_score'] = chile['away_score'].astype('int64')
```

```python
resultados['resultados'] = np.where(resultados.loc[:,'home_score'] > resultados.loc[:,'away_score'],'Local', 'Empate')
resultados['resultados'] = np.where(resultados.loc[:,'home_score'] < resultados.loc[:,'away_score'],'Visita', 'Empate')
```

```python
df =  chile.copy()
```

```python
df
```

```python
df.loc[df['home_score'] > df['away_score'] , 'resultados'] = 'Local'
df.loc[df['home_score'] < df['away_score'] , 'resultados'] = 'Visita'
df.loc[df['home_score'] == df['away_score'] , 'resultados'] = 'Empate'

```

```python
df
```

```python
df.loc[(df['home_team'] =='Chile') & (df['resultados'] =='Local') , 'R'] = 'Victoria'
df.loc[(df['home_team'] =='Chile') & (df['resultados'] =='Visita') , 'R'] = 'Derrota'
df.loc[(df['home_team'] =='Chile') & (df['resultados'] =='Empate') , 'R'] = 'Empate'

df.loc[(df['away_team'] =='Chile') & (df['resultados'] =='Local') , 'R'] = 'Derrota'
df.loc[(df['away_team'] =='Chile') & (df['resultados'] =='Visita') , 'R'] = 'Victoria'
df.loc[(df['away_team'] =='Chile') & (df['resultados'] =='Empate') , 'R'] = 'Empate'

```
